```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib
import seaborn
import sklearn
from scipy import stats
import scipy
```

```python
data=pd.read_csv('D:\REPO_progetto\TesiBiomedica\ERP_SUB',header=None)
data.head(5)
```

```python
#elimino le prime due colonne 
data.drop(data.columns[0], axis=1, inplace=True)
data.drop(data.columns[0], axis=1, inplace=True)
data.head()
```

```python
#trasformo i dati in matrice numpy 
tab=np.array(data)
```

```python
#ottengo gli indici 
Adx=np.arange(0,59,2)
Bdx=np.arange(1,60,2)
```

```python
#ottengo i vettori animali e uomini
vett_animali=[tab[i]  for  i in Adx]
vett_uomini= [tab[i]  for  i in Bdx]
```

```python
len(vett_animali)
```

```python
#Calcolo la media 
animali=0
uomini=0

for  i in range(29):
    animali=animali+vett_animali[i]
    uomini=uomini+vett_uomini[i]
    
animali=animali/30
uomini=uomini/30

#plotto i risultati
plt.figure(figsize = (10, 5))
x=np.linspace(-100,1000,269)
plt.plot(x,animali)
plt.plot(x,uomini)
```

```python
animali_c=np.delete(animali,[np.concatenate((np.arange(0,26,1),np.arange(120,269,1)),axis=0)])
uomini_c=np.delete(uomini,[np.concatenate((np.arange(0,26,1),np.arange(120,269,1)),axis=0)])
x_c=np.delete(x,[np.concatenate((np.arange(0,26,1),np.arange(120,269,1)),axis=0)])

plt.figure(figsize = (10, 5))
plt.plot(x_c,animali_c,color='k', label = 'animali')
plt.plot(x_c,uomini_c,color='r',label = 'uomini')
plt.axvline(x = 50, color = 'g')
plt.axvline(x = 100, color = 'g')
plt.axvline(x = 150, color = 'g')
plt.axvline(x = 200, color = 'g')
plt.axvline(x = 250, color = 'g')
plt.axvline(x = 300, color = 'g')
plt.axvline(x = 350, color = 'g')
plt.xlabel('Time (ms)')
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
```

```python
#Indici che definiscono gli intervalli 
Sid=[24,37,49,61,73,85,98,110,123]

#Ottengo gli intervalli dai vettori tempo e categorie animali e uomini 
S_x=[np.take(x,np.arange(Sid[i],Sid[i+1]))for  i in range(len(Sid)-1)]
S_a_m=[np.take(animali,np.arange(Sid[i],Sid[i+1]))for  i in range(len(Sid)-1)]
S_u_m=[np.take(uomini,np.arange(Sid[i],Sid[i+1]))for  i in range(len(Sid)-1)]

#ottengo i valori medi per ogni intervallo dei vettori delle categorie animali e uomini 
S_mean_a_m=[np.mean(i) for  i in S_a_m]
S_mean_u_m=[np.mean(i) for  i in S_u_m]

#plotto i risutati 
plt.figure(figsize = (10, 5))
plt.plot(range(8),S_mean_a_m,color='k', label = 'animali')
plt.plot(range(8),S_mean_u_m,color='r',label = 'uomini')
plt.xlabel('50 ms intervals')
plt.ylabel('mean of normalized data')
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
```

```python
#eseguo la segmentazione per tutti gli nimali e per tutti gli uomini 
S_a=[[np.take(j,np.arange(Sid[i],Sid[i+1]))for  i in range(len(Sid)-1)]for  j in vett_animali]
S_u=[[np.take(j,np.arange(Sid[i],Sid[i+1]))for  i in range(len(Sid)-1)] for  j in vett_uomini]


#eseguo la media e std per tutti
S_mean_a=[[np.mean(i) for  i in S_a[j]]for j in range(len(vett_animali))]
S_mean_u=[[np.mean(i) for  i in S_u[j]]for j in range(len(vett_animali))]
S_std_a=[[np.std(i) for  i in S_a[j]]for j in range(len(vett_animali))]
S_std_u=[[np.std(i) for  i in S_u[j]]for j in range(len(vett_animali))]
S_ent_a=[[scipy.stats.entropy(i, qk=None, base=None, axis=0) for  i in S_a[j]]for j in range(len(vett_animali))]
S_ent_u=[[scipy.stats.entropy(i, qk=None, base=None, axis=0) for  i in S_u[j]]for j in range(len(vett_animali))]
```

```python
plt.figure(figsize = (10, 5))
for  i in range(len(vett_animali)):
    plt.scatter(range(8),S_mean_u[i], color='r')
    
plt.figure(figsize = (10, 5))
for  i in range(len(vett_animali)):
    plt.scatter(range(8),S_mean_a[i], color='k' )
    
```

```python
plt.figure(figsize = (10, 5))
for  i in range(len(vett_animali)):
    plt.scatter(range(8),S_std_u[i], color='r')
    
    
plt.figure(figsize = (10, 5))
for  i in range(len(vett_animali)):
    plt.scatter(range(8),S_std_a[i], color='k')
```

```python
plt.figure(figsize = (10, 5))
for  i in range(len(vett_animali)):
    plt.scatter(range(8),S_ent_u[i], color='r')
    
    
plt.figure(figsize = (10, 5))
for  i in range(len(vett_animali)):
    plt.scatter(range(8),S_ent_a[i], color='k')
```

```python
from sklearn import svm
X = [S_mean_a_m,S_mean_u_m] 
y = [0, 1]
clf = svm.SVC()
clf.fit(X, y)
```

```python
pred_a=clf.predict(S_mean_a)
pred_u=clf.predict(S_mean_u)
```

```python
a=0;
u=0;
for i in range(30):
    if (pred_a[i]==0):
        a=a+1;
    if (pred_u[i]==1):
        u=u+1;
acc_a=a/30;
acc_u=u/30;
```
