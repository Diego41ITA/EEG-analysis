```python
#creo un array contenente i nomi dei file fif da leggere
nomi= ['epochs_1-epo.fif', 'epochs_2-epo.fif', 'epochs_4-epo.fif', 'epochs_5-epo.fif',
        'epochs_9-epo.fif', 'epochs_10-epo.fif', 'epochs_11-epo.fif', 'epochs_13-epo.fif', 
        'epochs_14-epo.fif', 'epochs_16-epo.fif', 'epochs_19-epo.fif', 'epochs_21-epo.fif',
        'epochs_24-epo.fif', 'epochs_26-epo.fif', 'epochs_27-epo.fif', 'epochs_29-epo.fif', 
        'epochs_30-epo.fif', 'epochs_33-epo.fif', 'epochs_34-epo.fif', 'epochs_35-epo.fif', 
        'epochs_36-epo.fif', 'epochs_37-epo.fif', 'epochs_38-epo.fif', 'epochs_39-epo.fif', 
        'epochs_40-epo.fif', 'epochs_42-epo.fif', 'epochs_45-epo.fif', 'epochs_46-epo.fif',
        'epochs_47-epo.fif']
print(len(nomi))

#leggo le epochs dai file fif
epochs_vector=[]
epochs_vector=[mne.read_epochs(i, proj=False, preload=True, verbose=None)for i in nomi]
print(len(epochs_vector))

#applico baseline alle epochs
for i in range (len(nomi)):
    epochs_vector[i].apply_baseline((None, 0))
```

```python
#vettori di evoked, uno spazio per oggetto 
animali_vector=[epochs_vector[i]['10002'].average()  for i in range(len(epochs_vector))]
umani_vector  =[epochs_vector[i]['10004'].average()  for i in range(len(epochs_vector))]
cibo_vector=[epochs_vector[i]['10003'].average()  for i in range(len(epochs_vector))]
veicoli_vector=[epochs_vector[i]['10005'].average()  for i in range(len(epochs_vector))]
strumenti_vector=[epochs_vector[i]['10006'].average()  for i in range(len(epochs_vector))]
print(len(animali_vector))
```

```python
#selezione dei canali attorno a Oz e Poz (rilevanti per discriminazione stimoli visivi)
roi_dict = dict(att_Oz=[17,16,15],att_Poz=[45,46,48,47]) 

#ottengo vettori di evoked per quanto riguarda la ROI (utilizzando la media sincrona)
roi_evoked_animali = [mne.channels.combine_channels(i, roi_dict, method='mean')for i in animali_vector]
roi_evoked_umani = [mne.channels.combine_channels(i, roi_dict, method='mean')for i in umani_vector]
roi_evoked_cibo = [mne.channels.combine_channels(i, roi_dict, method='mean')for i in cibo_vector]
roi_evoked_veicoli = [mne.channels.combine_channels(i, roi_dict, method='mean')for i in veicoli_vector]
roi_evoked_strumenti = [mne.channels.combine_channels(i, roi_dict, method='mean')for i in strumenti_vector]
```

```python
#estrapolazione dati dai vettori evoked 
#-->[soggetto][n ROI,samples]
ERP_ROI_animali_vector = [(roi_evoked_animali[i].data )for  i in range(len(roi_evoked_animali))]
ERP_ROI_umani_vector = [(roi_evoked_umani[i].data)   for  i in range(len(roi_evoked_umani))] 
ERP_ROI_cibo_vector = [(roi_evoked_cibo[i].data )for  i in range(len(roi_evoked_cibo))]
ERP_ROI_veicoli_vector = [(roi_evoked_veicoli[i].data)   for  i in range(len(roi_evoked_veicoli))] 
ERP_ROI_strumenti_vector = [(roi_evoked_strumenti[i].data )for  i in range(len(roi_evoked_strumenti))] 
```

```python
#verifica della dimensionalità
print(ERP_ROI_animali_vector[0].shape)
print(ERP_ROI_umani_vector[0].shape)
print(ERP_ROI_cibo_vector[0].shape)
print(ERP_ROI_veicoli_vector[0].shape)
print(ERP_ROI_strumenti_vector[0].shape)
print(ERP_ROI_animali_vector[1].shape)
print(ERP_ROI_umani_vector[1].shape)
print(ERP_ROI_cibo_vector[1].shape)
print(ERP_ROI_veicoli_vector[1].shape)
print(ERP_ROI_strumenti_vector[1].shape)
```

```python
#definizione funzione media sincrona
def moving_average(a, n=8) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
#applicazione media sincrona
ERP_ROI_Animali_av_Oz =[moving_average(ERP_ROI_animali_vector[i][0,:])for i in range(len(ERP_ROI_animali_vector))]
ERP_ROI_Animali_av_POz=[moving_average(ERP_ROI_animali_vector[i][1,:])for i in range(len(ERP_ROI_animali_vector))]

ERP_ROI_Uomini_av_Oz =[moving_average(ERP_ROI_umani_vector[i][0,:])for i in range(len(ERP_ROI_animali_vector))]
ERP_ROI_Uomini_av_POz=[moving_average(ERP_ROI_umani_vector [i][1,:])for i in range(len(ERP_ROI_animali_vector))]

ERP_ROI_Cibo_av_Oz=[moving_average(ERP_ROI_cibo_vector [i][0,:])for i in range(len(ERP_ROI_animali_vector))]
ERP_ROI_Cibo_av_POz=[moving_average(ERP_ROI_cibo_vector [i][1,:])for i in range(len(ERP_ROI_animali_vector))]

ERP_ROI_Veicoli_av_Oz=[moving_average(ERP_ROI_veicoli_vector [i][0,:])for i in range(len(ERP_ROI_animali_vector))]
ERP_ROI_Veicoli_av_POz=[moving_average(ERP_ROI_veicoli_vector [i][1,:])for i in range(len(ERP_ROI_animali_vector))]

ERP_ROI_Strumenti_av_Oz=[moving_average(ERP_ROI_strumenti_vector [i][0,:])for i in range(len(ERP_ROI_animali_vector))]
ERP_ROI_Strumenti_av_POz=[moving_average(ERP_ROI_strumenti_vector [i][1,:])for i in range(len(ERP_ROI_animali_vector))]
```

```python
#media matematica
ERP_ROI_Animali_vector=[((ERP_ROI_Animali_av_Oz[i]+ ERP_ROI_Animali_av_POz[i]) / 2)for i in range(len(ERP_ROI_Animali_av_POz))]
ERP_ROI_Uomini_vector =[((ERP_ROI_Uomini_av_Oz[i] + ERP_ROI_Uomini_av_POz[i])  / 2)for i in range(len(ERP_ROI_Uomini_av_POz))]      
ERP_ROI_Cibo_vector=[((ERP_ROI_Cibo_av_Oz[i]+ ERP_ROI_Cibo_av_POz[i]) / 2)for i in range(len(ERP_ROI_Cibo_av_POz))]
ERP_ROI_Veicoli_vector =[((ERP_ROI_Veicoli_av_Oz[i] + ERP_ROI_Veicoli_av_POz[i])  / 2)for i in range(len(ERP_ROI_Veicoli_av_POz))]      
ERP_ROI_Strumenti_vector =[((ERP_ROI_Strumenti_av_Oz[i] + ERP_ROI_Strumenti_av_POz[i])  / 2)for i in range(len(ERP_ROI_Strumenti_av_POz))]      
```

```python
#stampa ERP per usarli per feature extraction
for  i in range(len(ERP_ROI_Animali_vector)):
    ERP_ROI_Animali_vector[i]=ERP_ROI_Animali_vector[i].tolist()
    ERP_ROI_Animali_vector[i].insert(0,nomi[i])
    ERP_ROI_Animali_vector[i].insert(1,'Animali')
                                     
    ERP_ROI_Uomini_vector[i]=ERP_ROI_Uomini_vector[i].tolist()
    ERP_ROI_Uomini_vector[i].insert(0,nomi[i])
    ERP_ROI_Uomini_vector[i].insert(1,'Uomini')
    
    ERP_ROI_Cibo_vector[i]=ERP_ROI_Cibo_vector[i].tolist()
    ERP_ROI_Cibo_vector[i].insert(0,nomi[i])
    ERP_ROI_Cibo_vector[i].insert(1,'Cibo')
                                     
    ERP_ROI_Veicoli_vector[i]=ERP_ROI_Veicoli_vector[i].tolist()
    ERP_ROI_Veicoli_vector[i].insert(0,nomi[i])
    ERP_ROI_Veicoli_vector[i].insert(1,'Veicoli')
    
    ERP_ROI_Strumenti_vector[i]=ERP_ROI_Strumenti_vector[i].tolist()
    ERP_ROI_Strumenti_vector[i].insert(0,nomi[i])
    ERP_ROI_Strumenti_vector[i].insert(1,'Strumenti')




f=open("ERP_SUB","a",newline="")
writer=csv.writer(f)

for i in range(len(ERP_ROI_Animali_vector)):
    writer.writerow(ERP_ROI_Animali_vector[i])
    writer.writerow(ERP_ROI_Uomini_vector[i])
    writer.writerow(ERP_ROI_Cibo_vector[i])
    writer.writerow(ERP_ROI_Veicoli_vector[i])
    writer.writerow(ERP_ROI_Strumenti_vector[i])
    
f.close()
```

```python
#normalizzazione

for i in range(len(ERP_ROI_Animali_vector)):
    ERP_ROI_Animali_vector[i] = ERP_ROI_Animali_vector[i] / np.max(np.abs(ERP_ROI_Animali_vector[i]))
    ERP_ROI_Uomini_vector[i] = ERP_ROI_Uomini_vector[i] / np.max(np.abs(ERP_ROI_Uomini_vector[i]))
    ERP_ROI_Cibo_vector[i] = ERP_ROI_Cibo_vector[i] / np.max(np.abs(ERP_ROI_Cibo_vector[i]))
    ERP_ROI_Veicoli_vector[i] = ERP_ROI_Veicoli_vector[i] / np.max(np.abs(ERP_ROI_Veicoli_vector[i]))
    ERP_ROI_Strumenti_vector[i] = ERP_ROI_Strumenti_vector[i] / np.max(np.abs(ERP_ROI_Strumenti_vector[i]))
```

```python
#creo le funzioni per calcolare le features del segnale nel tempo

from scipy import stats
def mean(x):
    return np.mean(x,axis=-1)
def std(x):
    return np.std(x,axis=-1)
def ptp(x):
    return np.ptp(x,axis=-1)
def var(x):
    return np.var(x,axis=-1)
def minim(x):
    return np.min(x,axis=-1)
def maxim(x):
    return np.max(x,axis=-1)

def argminim(x):
    return np.argmin(x,axis=-1)

def argmaxim(x):
    return np.argmax(x,axis=-1)

def rms(x):
    return np.sqrt(np.mean(x**2,axis=-1))

def abs_diff_signal(x):
    return np.sum(np.abs(np.diff(x,axis=-1)),axis=-1)

def skewness(x):
    return stats.skew(x,axis=-1)
def kurtosis(x):
    return stats.kurtosis(x,axis=-1)

def concatenate_features(x):
    return np.concatenate((mean(x),std(x),ptp(x),var(x),minim(x),maxim(x),argminim(x),
    argmaxim(x),abs_diff_signal(x),skewness(x),kurtosis(x)),axis=-1)
```

```python
#calcolo e concateno le features
features_animali=[]
features_animali.append(concatenate_features(ERP_ROI_Animali_vector))
features_uomini=[]
features_uomini.append(concatenate_features(ERP_ROI_Uomini_vector))
```

```python
#ridimensiono gli array per la corretta stampa nei tsv
features_array_stamp_animali=np.array(features_animali)
features_array_stamp_animali=features_array_stamp_animali.reshape(30,11)
features_array_stamp_animali

features_array_stamp_uomini=np.array(features_uomini)
features_array_stamp_uomini=features_array_stamp_uomini.reshape(30,11)
features_array_stamp_uomini
```
